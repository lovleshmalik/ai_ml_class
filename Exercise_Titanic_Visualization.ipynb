```python
# Line Plot:
#    * Task 1: Create a line plot showing the trend of average age of passengers based on their travel class (1st, 2nd, and 3rd).
# 
#    * Task 2: Create a line plot showing the trend of survival rate of passengers based on their age group (child, adult, elderly).
```

```python
# Create a line plot showing the trend of average age of passengers based on their travel class (1st, 2nd, and 3rd).

plot1_df=pd.DataFrame({"AvgAge":[df[df["Pclass"]==1]["Age"].mean(),df[df["Pclass"]==2]["Age"].mean(),df[df["Pclass"]==3]["Age"].mean()],
                       "Pclass":[1,2,3]})
plot1_df
```

```python
sns.lineplot(x=plot1_df["Pclass"],y=plot1_df["AvgAge"])
```

```python
# Create a line plot showing the trend of survival rate of passengers based on their age group (child, adult, elderly).
```

```python
plot2_df=pd.DataFrame({"SurvivalRate":[df[df["Age"]<18]["Survived"].sum(),df[(df["Age"]>=18) & (df["Age"]<60)]["Survived"].sum(),df[df["Age"]>60]["Survived"].sum()],
                       "Category":["Child","Adult","Elderly"]})
plot2_df

# df[(df["Age"]>=18) & (df["Age"]<60)]["Survived"].sum()
```

```python
sns.lineplot(x=plot2_df["Category"],y=plot2_df["SurvivalRate"])
```

```python
# Scatter Plot:
#    * Task 1: Visualize the relationship between age and fare paid by passengers. Additionally, differentiate the points by their survival status (survived vs. not survived).
# 
#    * Task 2: Create a scatter plot showing the relationship between age and fare paid by passengers, differentiated by their gender (male vs. female).
```

```python
#Visualize the relationship between age and fare paid by passengers
sns.scatterplot(x=df["Age"],y=df["Fare"])
plt.title("Fare by Age")
```

```python
#Additionally, differentiate the points by their survival status (survived vs. not survived).
sns.scatterplot(x=df["Age"],y=df["Fare"],hue=df["Survived"])
plt.title("Fare by Age (with survival based segregation)")
```

```python
# Create a scatter plot showing the relationship between age and fare paid by passengers, differentiated by their gender (male vs. female).
```

```python
sns.scatterplot(x=df["Age"],y=df["Fare"],hue=df["Sex"])
plt.title("Fare by Age (with gender based segregation)")
```

```python
# Bar Plot:
#    * Task 1: Plot a bar plot showing the count of passengers who survived and who did not survive based on their gender (male vs. female).
```

```python
survived_by_gender_df=pd.DataFrame({"Survival_Count":[df[df["Sex"]=="male"]["Survived"].sum(),df[df["Sex"]=="female"]["Survived"].sum()],
                        "Gender":["Male","Female"]})
survived_by_gender_df
```

```python
sns.barplot(x=survived_by_gender_df["Gender"],y=survived_by_gender_df["Survival_Count"])
plt.title("Survivals by Gender")
```

```python
# Task 2: Plot a bar plot showing the count of passengers who survived and who did not survive based on their travel class (1st, 2nd, and 3rd).
```

```python
survived_by_pclass_df=pd.DataFrame({"Survival_Count":[df[df["Pclass"]==1]["Survived"].sum(),df[df["Pclass"]==2]["Survived"].sum(),df[df["Pclass"]==3]["Survived"].sum()],
                        "Class":["First","Second","Third"]})
survived_by_pclass_df
```

```python
sns.barplot(x=survived_by_pclass_df["Class"],y=survived_by_pclass_df["Survival_Count"])
plt.title("Survivals by Class")
```

```python
# Box Plot:
#    * Task 1: Visualize the distribution of fares paid by passengers based on their embarkation port (Cherbourg, Queenstown, Southampton).
#   
#    * Task 2: Visualize the distribution of ages of passengers based on their survival status (survived vs. not survived).
```

```python
#    * Task 1: Visualize the distribution of fares paid by passengers based on their embarkation port (Cherbourg, Queenstown, Southampton).
```

```python
df[df["Embarked"]=="S"]["Fare"].values
# df["Embarked"].value_counts()
```

```python
fares_by_embarkation_point_df=pd.DataFrame({"Embarkation_Point":["Southampton","Queenstown","Cherbourg"],
                                            "Fares":[
                                                df[df["Embarked"]=="S"]["Fare"].values,df[df["Embarked"]=="Q"]["Fare"].values,df[df["Embarked"]=="C"]["Fare"].values
                                            ]})
fares_by_embarkation_point_df
```

```python
# sns.boxplot(x=fares_by_embarkation_point_df["Embarkation_Point"], y=fares_by_embarkation_point_df["Fares"])
# fares_by_embarkation_point_df['Fares'].apply(lambda x:pd.Series(x).describe())
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(fares_by_embarkation_point_df['Fares'],labels=fares_by_embarkation_point_df['Embarkation_Point'])
plt.title("Fares by Embarkation point")
plt.xlabel("Embarkation_Point")
plt.ylabel("Fares")
plt.grid(True)
plt.show()
```

```python
# Task 2: Visualize the distribution of ages of passengers based on their survival status (survived vs. not survived).
```

```python
df_ignore_nan=df.dropna()
age_by_survival_rate_df=pd.DataFrame({"SurvivalState":["Survived","Not Survived"],
                                            "Age":[
                                                df_ignore_nan[df_ignore_nan["Survived"]==1]["Age"].values,df_ignore_nan[df_ignore_nan["Survived"]==0]["Age"].values
                                            ]})
age_by_survival_rate_df
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(age_by_survival_rate_df['Age'],labels=age_by_survival_rate_df['SurvivalState'])
# plt.title("Age by Survival Rate")
plt.xlabel("Survival State")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```
